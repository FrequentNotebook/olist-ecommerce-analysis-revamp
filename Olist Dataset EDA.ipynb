```python
# import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def load_olist_data(file_map):
    
    loaded_data = {}
    
    for name, path in file_map.items():
        
        if not os.path.exists(path):
            print(f'{name} file path does not exist!')
            continue
            
        try: 
            loaded_data[name] = pd.read_csv(path, low_memory=False)
        except Exception as e:
            print(f'Error loading {name}: {e}')
            
    return loaded_data

files = {
    'customers' : 'cleaned_customers.csv',
    'geolocation' : 'cleaned_geolocation.csv',
    'order_items' : 'cleaned_order_items.csv',
    'orders' : 'cleaned_orders.csv',
    'payments': 'cleaned_payment.csv',
    'product_category' : 'cleaned_product_category.csv',
    'products' : 'cleaned_products.csv',
    'reviews' : 'cleaned_reviews.csv',
    'sellers' : 'cleaned_sellers.csv'
}

            
data = load_olist_data(files)
```

```python
def clean_and_validate(data_dict):
    """
    This function performs an audit on all dataframes, converts types, 
    and validates integrity.
    """
    for name, df in data_dict.items():
        for col in df.columns:
            
            # Convert and Validate Dates
            if 'date' in col or 'timestamp' in col:
                df[col] = pd.to_datetime(df[col], errors='coerce')
                is_datetime = df[col].dtype.kind == 'M' 
                assert is_datetime, f"{name}['{col}'] failed datetime conversion"
                
            # Convert and Validate Zip Codes
            if 'zip_code_prefix' in col:
                df[col] = df[col].astype(str).str.zfill(5)
                is_object = df[col].dtype == 'object'
                assert is_object, f"{name}['{col}'] failed string conversion"
                
        print(f'{name} is cleaned and verified')
        
    return data_dict

data = clean_and_validate(data)
```

## Orders

Note: we have a `is_imputed_approved_at` column which was engineered for orders that was delivered but did not have an `order_approved_at` column. The fields were filled with `order_purchase_timestamp` values. You can check Notebook 1 for more information or how it was done.

```python
data['orders']
```

```python
# check the order_status percentage of all the orders

status_pct = data['orders']['order_status'].value_counts(normalize=True) * 100
status_pct.round(2)
```

```python
# check the percentage of the imputed approved orders

imputed_pct = data['orders']['is_imputed_approved_at'].mean()*100
print(f"imputed rate: {imputed_pct:.3f}%")
```

```python
# select delivered orders without imputed data

delivered_df = data['orders'][
    (data['orders']['order_status'] == 'delivered') & 
    (data['orders']['is_imputed_approved_at'] == 0)
].copy()

# Was it approved after purchase?
delivered_df['order_logic_1'] = delivered_df['order_approved_at'] >= delivered_df['order_purchase_timestamp']

# Was it handed to carrier after approval?
delivered_df['order_logic_2'] = delivered_df['order_delivered_carrier_date'] >= delivered_df['order_approved_at']

# Was it delivered to customer after carrier pick-up?
delivered_df['order_logic_3'] = delivered_df['order_delivered_customer_date'] >= delivered_df['order_delivered_carrier_date']

# Overall Symmetry: Did it pass all three?
delivered_df['is_linear'] = delivered_df['order_logic_1'] & delivered_df['order_logic_2'] & delivered_df['order_logic_3']

# Summary of results
print(f"Total Delivered Orders Checked: {len(delivered_df)}")
print(f"Linear Orders: {delivered_df['is_linear'].sum()}")
print(f"Non-Linear Orders: {len(delivered_df) - delivered_df['is_linear'].sum()}")
```

```python
# Let's see which step failed most often
step_1_fails = (~delivered_df['order_logic_1']).sum() # Approved before Purchase
step_2_fails = (~delivered_df['order_logic_2']).sum() # Carrier before Approved
step_3_fails = (~delivered_df['order_logic_3']).sum() # Customer before Carrier

print(f" Logic 1 Fails: {step_1_fails}")
print(f" Logic 2 Fails: {step_2_fails}")
print(f" Logic 3 Fails: {step_3_fails}")
```

```python
# feature engineer a linear order column

logic_pass = (
    (data['orders']['order_status'] == 'delivered') &
    (data['orders']['is_imputed_approved_at'] == 0) &
    (data['orders']['order_approved_at'] >= data['orders']['order_purchase_timestamp']) &
    (data['orders']['order_delivered_carrier_date'] >= data['orders']['order_approved_at']) &
    (data['orders']['order_delivered_customer_date'] >= data['orders']['order_delivered_carrier_date'])
)


data['orders']['is_linear_timestamp'] = logic_pass.fillna(False)

print(data['orders']['is_linear_timestamp'].value_counts())
```
